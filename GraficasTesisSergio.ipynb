```python
import pandas as pd
import plotly.express as px
```

```python
GLB = pd.read_csv('GLB.csv')
MET =  pd.read_csv('MET.csv') 
GI0C1 = pd.read_csv('GI0C1.csv')
```

```python
print(len(GLB))
print(len(MET))
print(len(GI0C1))
```

```python
GLB
```

```python
MET.reset_index(inplace = True, drop = False)
```

```python
MET
```

```python
MET1 = MET.drop(['level_0'], axis=1)
MET1
```

```python
ME = MET1.set_axis(['index', 'rMET', 'cMET', 'nombreMET'], axis=1)
```

```python
GLB.reset_index(inplace = True, drop = False)
```

```python
GLB1 = GLB.drop(['level_0'], axis=1)
GLB1
```

```python
GI0C1.reset_index(inplace = True, drop = False)
```

```python
C1 = GI0C1.drop(['level_0'], axis=1)
C1
```

```python
GL = GLB1.set_axis(['index', 'rGLB', 'cGLB', 'nombreGLB'], axis=1)
```

```python
C = C1.set_axis(['index', 'rGI0C1', 'cGI0C1', 'nombreGI0C1'], axis=1)
```

```python
GL
```

```python
Union1 = pd.merge(C,GL,how='left', left_on=['index'],right_on=['index'])
Union1
```

```python
Union2 = pd.merge(Union1,ME,how='left', left_on=['index'],right_on=['index']) 
Union2
```

```python
Union2.to_csv('ResumenDatos.csv')
```

```python
base = pd.concat([GLB, MET, GI0C1], axis=0)
base
```

```python
graf = px.line(base, x = 'r', y = 'c', title=' GLB, MET & CI0C1', color = 'nombre')
graf.show()
```

```python
base2 = pd.concat([MET, GI0C1], axis=0)
base2
```

```python
graf2 = px.line(base2, x = 'r', y = 'c', title='MET & CI0C1', color = 'nombre')
graf2.show()
```

```python
base3 = pd.concat([MET, GLB], axis=0)
base3
```

```python
graf3 = px.line(base3, x = 'r', y = 'c', title='MET & CI0C1', color = 'nombre')
graf3.show()
```
