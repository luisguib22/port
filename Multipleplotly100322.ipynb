```python
import pandas as pd
import plotly.express as px
btc = pd.read_csv('Bitcoin USD (BTC-USD)2022-03-10.csv')
meta = pd.read_csv('Meta Platforms, Inc.  (FB)2022-03-10.csv')
micro = pd.read_csv('Microsoft Corporation (MSFT)2022-03-10.csv')
```

```python
base = pd.concat([meta, btc, micro], axis=0)
base
```

```python
dos_lineas = px.line(base, x = 'Fecha', y = 'Cierre', title=' Meta Platforms, Inc & BTC-USD, Microsoft', color = 'Nombre')
dos_lineas.show()
```
